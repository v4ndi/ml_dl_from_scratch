# Testing custom metrics

```python
import numpy as np
import sklearn.metrics
import matplotlib.pyplot as plt
import metrics
```

# Regression metrics

## MSE

```python
y_true = np.random.rand(100)
y_pred  = np.random.rand(100)

print(f'sklearn MSE: {sklearn.metrics.mean_squared_error(y_true, y_pred)}')
print(f'mse: {metrics.mse(y_true, y_pred)}')
```

## MAE

```python
print(f'sklearn MAE: {sklearn.metrics.mean_absolute_error(y_true, y_pred)}')
print(f'MAE: {metrics.mae(y_true, y_pred)}')
```

## R2_score

```python
print(f'sklearn R2 score: {sklearn.metrics.r2_score(y_true, y_pred)}')
print(f'R2 score: {metrics.r2_score(y_true, y_pred)}')
```

## Classification metrics

## Confussion matrix

```python
y_true = np.random.randint(0, 5, 100)
y_pred = np.random.randint(0, 5, 100)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
conf_matrix_sklearn = sklearn.metrics.confusion_matrix(y_true, y_pred)
conf_matrix = metrics.confussion_matrix(y_true, y_pred)
ConfusionMatrixDisplay(conf_matrix_sklearn).plot()
ConfusionMatrixDisplay(conf_matrix).plot()
```

## ROC-curve and ROC-AUC score

```python
y_true = np.random.randint(0, 2, 100)
y_pred = np.random.uniform(0, 1, 100)

FPR_sk, TPR_sk, _ = sklearn.metrics.roc_curve(y_true, y_pred)
FPR, TPR = metrics.roc_curve(y_true, y_pred)

plt.plot(FPR_sk, TPR_sk)
```

```python
plt.plot(FPR, TPR)
```

```python
print('sklearn roc-auc score', sklearn.metrics.roc_auc_score(y_true, y_pred))
print('custom roc-auc score', metrics.roc_auc_score(y_true, y_pred))
```

## Accuracy

```python
y_true = np.random.randint(0, 5, 100)
y_pred = np.random.randint(0, 5, 100)

print('sklearn Accuracy', sklearn.metrics.accuracy_score(y_true, y_pred))
print('custom Accuracy', metrics.accuracy_score(y_true, y_pred))
```

## Precision

#### Binary

```python
y_true = np.random.randint(0, 2, 100)
y_pred = np.random.randint(0, 2, 100)

print('sklearn precission score', sklearn.metrics.precision_score(y_true, y_pred))
print('custom precission score', metrics.precision_score(y_true, y_pred))
```

#### Micro

```python
y_true = np.random.randint(0, 5, 100)
y_pred = np.random.randint(0, 5, 100)

print('skleran precision score ',sklearn.metrics.precision_score(y_true, y_pred, average='micro'))
print('custom precision score',metrics.precision_score(y_true, y_pred, average='micro'))
```

#### Macro

```python
print('skleran precision score', sklearn.metrics.precision_score(y_true, y_pred, average='macro'))
print('skleran precision score', metrics.precision_score(y_true, y_pred, average='macro'))
```
